## 1. `print()` e comentários

- `print()` é o comando mais básico do Python: ele **mostra algo na tela**.
- Comentários são anotações no código que o Python **ignora** ao executar. Servem só para humanos lerem. Começam com `#` (uma linha) ou ficam entre `"""` (várias linhas).

```python
# Isso é um comentário — o Python não executa esta linha
print("Olá, mundo!")  # isso aqui também é um comentário, depois do código

"""
Isso é um comentário
de várias linhas.
Útil para explicações maiores.
"""

print("Aprendendo Python é mais fácil do que parece.")
```

## 2. Variáveis e tipos de dados

Uma **variável** é como uma caixa com etiqueta: você guarda um valor nela e usa o nome para acessá-lo depois.

Os tipos mais comuns em Python:

| Tipo | O que é | Exemplo |
|---|---|---|
| `str` | texto (string) | `"João"` |
| `int` | número inteiro | `30` |
| `float` | número com casas decimais | `75.5` |
| `bool` | verdadeiro ou falso | `True` / `False` |

Use `type(variavel)` para descobrir o tipo de algo.

```python
nome = "João"
idade = 30
peso = 75.5
esta_matriculado = True

print(nome, type(nome))
print(idade, type(idade))
print(peso, type(peso))
print(esta_matriculado, type(esta_matriculado))
```

💡 **Dica:** em Python você não precisa "avisar" o tipo da variável — ele descobre sozinho. E por convenção, nomes de variáveis usam `snake_case` (palavras separadas por `_`), tudo minúsculo. Para constantes (valores que nunca mudam), usa-se `MAIUSCULO`.

```python
VARIAVEL_CONSTANTE = "Isso não deveria mudar durante o programa"
print(VARIAVEL_CONSTANTE)
```

## 3. Entrada de dados com `input()`

Até agora usamos valores fixos. Com `input()`, o programa **pergunta algo ao usuário** e guarda a resposta digitada — sempre como texto (`str`).

Se você precisa de um número, converte o resultado com `int(...)` ou `float(...)`.

```python
nome = input("Digite seu nome: ")
print(f"Olá, {nome}! Seja bem-vindo(a).")

idade_texto = input("Digite sua idade: ")
idade = int(idade_texto)  # convertendo texto para número inteiro
print(f"Daqui a 10 anos você terá {idade + 10} anos.")
```

💡 Repare no `f"..."` acima: é uma **f-string**, a forma mais moderna de montar textos com variáveis dentro. Tudo que estiver entre `{}` é calculado e inserido no texto.

```python
celsius = float(input("Digite a temperatura em Celsius: "))
fahrenheit = celsius * 9/5 + 32
print(f"{celsius}°C equivalem a {fahrenheit}°F")
```

## 4. Operadores

**Aritméticos** — para fazer contas:

| Operador | Significado | Exemplo |
|---|---|---|
| `+` | soma | `10 + 5` → `15` |
| `-` | subtração | `10 - 5` → `5` |
| `*` | multiplicação | `10 * 5` → `50` |
| `/` | divisão | `10 / 5` → `2.0` |
| `//` | divisão inteira | `10 // 3` → `3` |
| `%` | resto da divisão | `10 % 3` → `1` |
| `**` | potência | `10 ** 2` → `100` |

```python
a = 10
b = 3

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Divisão inteira:", a // b)
print("Resto (módulo):", a % b)
print("Potência:", a ** b)
```

**Relacionais** — comparam valores e sempre resultam em `True` ou `False`: `>`, `<`, `>=`, `<=`, `==` (igual), `!=` (diferente).

**Lógicos** — combinam condições: `and` (e), `or` (ou), `not` (nega).

```python
idade = 20
tem_carteira = True

print(idade >= 18)                     # relacional
print(idade >= 18 and tem_carteira)    # lógico: as duas precisam ser True
print(idade < 18 or tem_carteira)      # lógico: basta uma ser True
print(not tem_carteira)                # lógico: inverte o valor
```

```python
numero = 17
print(numero % 2 == 0)  # True se for par, False se for ímpar
```

## 5. Estruturas de dados

- **Lista** `[]` — coleção ordenada e **mutável** (pode mudar).
- **Tupla** `()` — coleção ordenada e **imutável** (não pode mudar).
- **Dicionário** `{chave: valor}` — guarda pares de chave e valor, como uma ficha de cadastro.
- **Set (conjunto)** `{}` — coleção sem ordem e **sem itens repetidos**.

```python
compras = ['maçã', 'banana', 'pão', 'leite']  # lista
print(compras)
compras.append('café')   # dá pra adicionar itens
print(compras)
print("Primeiro item:", compras[0])   # acessando por posição (começa em 0)
```

```python
endereco = {
    "rua": "Rua das Flores",
    "numero": 123,
    "cidade": "São Paulo",
}

print(endereco["cidade"])   # acessando valor pela chave
endereco["numero"] = 456    # alterando um valor
print(endereco)
```

```python
conjunto_a = {1, 2, 3, 4, 5}
conjunto_b = {3, 4, 5, 6, 7}

print(conjunto_a & conjunto_b)  # interseção: o que está nos dois
print(conjunto_a | conjunto_b)  # união: todos os elementos, sem repetir
```

## 6. Strings e seus métodos

Texto (`str`) tem vários métodos prontos para manipulação. Alguns dos mais usados:

```python
texto = "  Aprendo muito com o Python!  "

print(texto.strip())          # remove espaços das pontas
print(texto.upper())          # tudo maiúsculo
print(texto.lower())          # tudo minúsculo
print(texto.strip().replace("Python", "Django"))  # troca uma parte
print(texto.strip().split(" "))  # separa em uma lista de palavras
print(len(texto))             # tamanho do texto (incluindo espaços)
```

```python
frase = "não gosto de café"
print(frase.find("café"))          # posição onde "café" começa
print(frase[10:12])                # fatiamento: caracteres da posição 10 a 11
print(frase.startswith("não"))     # verifica início
print(frase.endswith("café"))      # verifica fim
```

```python
palavra = "arara"
print(palavra == palavra[::-1])  # [::-1] inverte o texto; compara com o original -> palíndromo?
```

## 7. Condicionais: `if`, `elif`, `else`

Servem para o programa **tomar decisões**: "se isso for verdade, faça aquilo; senão, faça outra coisa".

```python
idade = 15

if idade >= 18:
    print("Maior de idade")
elif idade >= 12:
    print("Adolescente")
else:
    print("Criança")
```

⚠️ Repare na **indentação** (espaços no início da linha): em Python, ela não é só estética — é o que define o que está "dentro" do `if`. Sempre use 4 espaços (ou um Tab) para indentar.

```python
ano = 2024

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"{ano} é bissexto")
else:
    print(f"{ano} não é bissexto")
```

## 8. Laços de repetição: `for` e `while`

- `for` repete um bloco de código **para cada item** de uma sequência (lista, texto, intervalo de números...).
- `while` repete um bloco de código **enquanto uma condição for verdadeira**.

```python
compras = ['maçã', 'banana', 'pão']

for item in compras:
    print("Item:", item)

print("---")

for numero in range(1, 6):   # de 1 até 5
    print("Número:", numero)
```

```python
contador = 0

while contador < 5:
    print("Contador vale:", contador)
    contador += 1   # o mesmo que: contador = contador + 1

print("Acabou o while!")
```

⚠️ **Cuidado com loops infinitos**: se a condição do `while` nunca virar `False`, o código roda para sempre. Sempre garanta que algo dentro do loop muda a condição.

```python
numero = 7
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
```

## 9. Funções

Uma função é um **bloco de código reutilizável**, que pode receber entradas (parâmetros) e devolver um resultado (`return`). Serve para não repetir o mesmo código várias vezes.

```python
def saudacao():
    print("Olá, mundo!")

def soma(a, b):
    return a + b

def eh_par(numero):
    return numero % 2 == 0

saudacao()
resultado = soma(10, 5)
print("Soma:", resultado)
print("7 é par?", eh_par(7))
```

💡 Parâmetros podem ter um **valor padrão**, usado quando você não informa nada:

```python
def saudacao_personalizada(nome="visitante"):
    print(f"Olá, {nome}!")

saudacao_personalizada()          # usa o padrão "visitante"
saudacao_personalizada("Maria")   # usa "Maria"
```

```python
def media(nota1, nota2, nota3):
    return (nota1 + nota2 + nota3) / 3

print(media(7, 8, 9))
```

## 10. Bibliotecas prontas: `random` e `math`

Python já vem com várias bibliotecas prontas para usar — basta importar com `import`. Duas muito úteis para começar:

- **`random`** — gera números aleatórios, sorteios, embaralhamentos.
- **`math`** — funções e constantes matemáticas (raiz quadrada, potência, arredondamento, `pi`, fatorial...).

Vamos ver exemplos de cada uma.

### `random`

| Função | O que faz |
|---|---|
| `random.randint(a, b)` | sorteia um inteiro entre `a` e `b` (incluindo os dois) |
| `random.choice(lista)` | escolhe um item aleatório da lista |
| `random.shuffle(lista)` | embaralha a lista (altera ela mesma) |
| `random.random()` | sorteia um número decimal entre 0 e 1 |

```python
import random

# Sorteando um número inteiro entre 1 e 10
numero_sorteado = random.randint(1, 10)
print("Número sorteado:", numero_sorteado)

# Escolhendo um item aleatório de uma lista
frutas = ['maçã', 'banana', 'uva', 'pêra', 'manga']
fruta_sorteada = random.choice(frutas)
print("Fruta sorteada:", fruta_sorteada)

# Embaralhando uma lista
baralho = ['A', 'K', 'Q', 'J', '10']
random.shuffle(baralho)
print("Baralho embaralhado:", baralho)
```

**Exemplo prático:** um jogo de adivinhação, onde o computador sorteia um número e o jogador tenta acertar.

```python
import random

numero_secreto = random.randint(1, 100)
tentativas = 0
acertou = False

print("Tente adivinhar o número entre 1 e 100!")

while not acertou:
    palpite = int(input("Seu palpite: "))
    tentativas += 1
    if palpite < numero_secreto:
        print("O número é maior!")
    elif palpite > numero_secreto:
        print("O número é menor!")
    else:
        acertou = True
        print(f"Parabéns! Você acertou em {tentativas} tentativas.")
```

### `math`

| Função / constante | O que faz |
|---|---|
| `math.sqrt(x)` | raiz quadrada de `x` |
| `math.pow(x, y)` | `x` elevado a `y` |
| `math.factorial(x)` | fatorial de `x` |
| `math.pi` | valor da constante π (3.14159...) |
| `math.ceil(x)` / `math.floor(x)` | arredonda para cima / para baixo |

```python
import math

print("Raiz quadrada de 81:", math.sqrt(81))
print("2 elevado a 10:", math.pow(2, 10))
print("Fatorial de 5:", math.factorial(5))
print("Valor de pi:", math.pi)
print("Arredondando 4.3 para cima:", math.ceil(4.3))
print("Arredondando 4.7 para baixo:", math.floor(4.7))
```

**Exemplo prático:** calculando a área de um círculo, usando `math.pi`.

```python
import math

raio = float(input("Digite o raio do círculo: "))
area = math.pi * (raio ** 2)
print(f"A área do círculo é: {area:.2f}")
```
